```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
file_path = 'D:/UNNES/Semester 4/Penambangan Data dan Analisis Bisnis/diabetes.csv'
data = pd.read_csv(file_path, delimiter=';')

# Ubah tipe data ke numerik jika memungkinkan
data = data.astype({
    'Pregnancies': 'int',
    'Glucose': 'int',
    'BloodPressure': 'int',
    'SkinThickness': 'int',
    'Insulin': 'int',
    'BMI': 'float',
    'DiabetesPedigreeFunction': 'float',
    'Age': 'int',
    'Outcome': 'int'
})
```

```python
# 1. Distribusi kadar Glukosa
plt.figure(figsize=(6, 4))
sns.histplot(data=data, x='Glucose', kde=True, color='skyblue')
plt.title('Distribusi Kadar Glukosa')
plt.show()
```

```python
# 2. Boxplot BMI berdasarkan Outcome
plt.figure(figsize=(6, 4))
sns.boxplot(data=data, x='Outcome', y='BMI', palette='Set2')
plt.title('Boxplot BMI berdasarkan Outcome')
plt.show()
```

```python
# 3. Scatterplot Usia vs Kadar Glukosa
plt.figure(figsize=(6, 4))
sns.scatterplot(data=data, x='Age', y='Glucose', hue='Outcome', palette='coolwarm')
plt.title('Scatterplot Usia vs Kadar Glukosa')
plt.show()
```

```python
# 4. Heatmap korelasi antar variabel
plt.figure(figsize=(8, 6))
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap Korelasi Antar Variabel')
plt.show()
```

```python
# 5. Countplot distribusi Outcome
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Outcome', palette='pastel')
plt.title('Distribusi Outcome')
plt.show()
```
